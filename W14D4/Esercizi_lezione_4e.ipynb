```python
# task 1:
'''Il dataset pokemon.csv contiene un database di Pokémon, 
con dati quali nome, tipi di attacco, valori di attacco/difesa/eccetera, e se sono o meno leggendari.
importa db'''

import pandas as pd

df = pd.read_csv('pokemon.csv')
```

```python
# task 2:
'''• Verifichiamo la dimensione, un'anteprima e osserviamo i nomi di colonna'''

df.shape
df.columns
```

```python
# task 4:
'''• Confrontiamolo con l'indice messo automaticamente da Pandas: combaciano?'''

# no
```

```python
# task 5:
'''• Se no, settare la prima colonna come indice.'''

df.set_index('#')
```

```python
df.drop('#', axis = 1)
```

```python
# task 6:
'''• Quali sono i Pokémon leggendari?'''

pokemon_legendari = df[df['Legendary'] == True]
pokemon_legendari
```

```python
# task 7:
'''• E quali sono i leggendari di tipo 1 Grass?'''

pokemon_legendari_grass = df[(df['Legendary'] == True) & (df['Type 1'] == 'Grass')]
pokemon_legendari_grass
```

```python
# task 8:
'''• E leggendari di tipo 1 Ice o Fire?'''

pokemon_legendari_ice_fire = df[(df['Legendary'] == True) & ((df['Type 1'] == 'Fire') | (df['Type 1'] == 'Ice'))]
pokemon_legendari_ice_fire
```

```python
# task 9:
'''• Ordiniamo il dataset per la colonna Name'''

df.sort_values(by = 'Name', ascending = True)
```

```python
# task 10:
'''• Trasformiamo Name nell'indice'''

df.set_index('Name', inplace = True)
```

```python
# task 11:
'''• Quali sono i Pokémon della prima generazione con attacco > 50 e HP < 60?'''

pokemon_1_gen = df[(df['Generation'] == 1) & (df['Attack'] > 50) & (df['HP'] < 60)]
pokemon_1_gen
```
