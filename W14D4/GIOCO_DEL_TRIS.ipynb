```python
# task BONUS:
'''CREARE IL GIOCO DEL TRIS'''

import random
import numpy as np

lista_numeri = np.random.randint(1, 3, 100)
lista_numeri = lista_numeri.astype(str)
lista_numeri = np.where(lista_numeri == '1', 'x', 'o')
lista_giocate = lista_numeri
iterazione = 1
contatore_x = 0
contatore_o = 0

for giocata in lista_giocate:
    print('giocata n°',iterazione,':',giocata)
    iterazione += 1
    if giocata == 'x':
        contatore_x += 1
    else: contatore_x = 0 
    if contatore_x == 3:
        print('il giocatore x vince!')
        break
    elif giocata == 'o':
        contatore_o += 1
    else: contatore_o = 0
    if contatore_o == 3:
        print('il giocatore o vince!')
        break
```

```python
import pandas as pd

# Definisci i nomi delle colonne e delle righe
colonne = ['A', 'B', 'C']
righe = ['1', '2', '3']

# Crea un DataFrame vuoto con valori iniziali come stringhe vuote
tabella = pd.DataFrame('', index=righe, columns=colonne)

# Mostra la tabella iniziale
print("Tabella iniziale:")
print(tabella)

# Variabile per alternare tra "X" e "O"
simbolo_corrente = 'X'

def check_vittoria(tabella, simbolo):
    # Controllo righe
    for riga in righe:
        if all(tabella.loc[riga, col] == simbolo for col in colonne):
            return True
    
    # Controllo colonne
    for col in colonne:
        if all(tabella.loc[riga, col] == simbolo for riga in righe):
            return True
    
    # Controllo diagonali
    if all(tabella.loc[righe[i], colonne[i]] == simbolo for i in range(3)) or \
       all(tabella.loc[righe[i], colonne[2-i]] == simbolo for i in range(3)):
        return True
    
    return False

# Ciclo continuo fino a riempire la tabella o trovare un vincitore
while True:
    # Mostra la tabella attuale
    print("\nTabella attuale:")
    print(tabella)

    # Chiedi all'utente di inserire la posizione
    posizione = input(f"Inserisci la posizione per '{simbolo_corrente}' (es. B2, o 'exit' per uscire): ")

    # Controllo per uscire dal ciclo
    if posizione.lower() == 'exit':
        break

    # Estrai la colonna e la riga dalla posizione fornita
    colonna = posizione[0].upper()  # estrae la colonna, e.g., 'B'
    riga = posizione[1]             # estrae la riga, e.g., '2'

    # Verifica se la posizione è valida e libera
    if colonna in colonne and riga in righe:
        if tabella.at[riga, colonna] == '':
            # Inserisci il simbolo nella posizione specificata
            tabella.at[riga, colonna] = simbolo_corrente

            # Controlla se c'è una vittoria
            if check_vittoria(tabella, simbolo_corrente):
                print(f"\nTabella finale:")
                print(tabella)
                print(f"\nHai vinto! '{simbolo_corrente}' ha completato una linea.")
                break

            # Alterna il simbolo per il prossimo turno
            simbolo_corrente = 'O' if simbolo_corrente == 'X' else 'X'
        else:
            print("Errore: la posizione è già occupata. Riprova.")
    else:
        print("Errore: posizione non valida. Riprova.")
```
