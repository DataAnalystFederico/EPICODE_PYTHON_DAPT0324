```python
# task 1:
'''Il dataset diabetes.csv raccoglie persone con diabete o meno, 
e il valore di diverse variabili fisiologiche dei pazienti.
importa db'''

import pandas as pd

df = pd.read_csv('diabetes.csv')
df
```

```python
# task 5:
'''• Selezioniamo i dati relativi a diverse fasce di età: 
<20, 20-30, 30-40, 40-50, >50'''

df[df['Age (years)'] < 20]
df[(df['Age (years)'] >= 20) & (df['Age (years)'] <= 30)]
df[(df['Age (years)'] >= 30) & (df['Age (years)'] <= 40)]
df[(df['Age (years)'] >= 40) & (df['Age (years)'] <= 50)]
df[df['Age (years)'] > 50]
```

```python
# task 6:
'''• Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?'''

pressione_media_under_20 = df['Diastolic blood pressure (mm Hg)'][df['Age (years)'] < 20].mean()
pressione_media_20_30 = df['Diastolic blood pressure (mm Hg)'][(df['Age (years)'] >= 20) & (df['Age (years)'] <= 30)].mean()
pressione_media_30_40 = df['Diastolic blood pressure (mm Hg)'][(df['Age (years)'] >= 30) & (df['Age (years)'] <= 40)].mean()
pressione_media_40_50 = df['Diastolic blood pressure (mm Hg)'][(df['Age (years)'] >= 40) & (df['Age (years)'] <= 50)].mean()
pressione_media_over_50 = df['Diastolic blood pressure (mm Hg)'][df['Age (years)'] > 50].mean()

print('la pressione media under 20 è:',pressione_media_under_20)
print('la pressione media 20-30 è:',pressione_media_20_30)
print('la pressione media 30-40 è:',pressione_media_30_40)
print('la pressione media 40-50 è:',pressione_media_40_50)
print('la pressione over 50 è:',pressione_media_over_50)
```

```python
# task 7:
'''• Qual è la media della pressione per ogni singolo anno di età?'''


df.groupby('Age (years)')['Diastolic blood pressure (mm Hg)'].mean()
```
