```python
# task 1:
'''importa db party_in_nyc e:

• Realizziamo un grafico che mostra quante chiamate ci sono
state per ogni città per tutti i tipi di locale'''

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('party_in_nyc.csv')


if df.isna().sum().sum() > 0:
    for column in df.columns:
        n_null = df[column].isna().sum()
        percentuale_null = round((n_null / df.shape[0]) * 100, 2)
        if percentuale_null > 0:
            print(column, 'contiene:', n_null,'dati null. Rappresentano il', percentuale_null,'% della colonna')
else:
    print('il db NON presenta dati null')
```

```python
df['Calls'] = df.index + 1

plt.figure(figsize=(20, 40))

sns.barplot( data = df, x = 'Calls', y = 'City', hue = 'Location Type')
```
