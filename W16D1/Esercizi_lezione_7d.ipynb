```python
# à task 1:
'''Importa db election e:

• Con un grafico a barre confrontiamo i voti totali presi dai tre candidati
(come somma di tutti i distretti)'''

import pandas as pd
import seaborn as sns
import matplotlib as plt

df = pd.read_csv('election.csv')

if df.isna().sum().sum() > 0:
    for column in df.columns:
        n_null = df[column].isna().sum()
        percentuale_null = round((n_null / df.shape[0]) * 100, 2)
        if percentuale_null > 0:
            print(column, 'contiene:', n_null,'dati null. Rappresentano il', percentuale_null,'% della colonna')
else:
    print('il db NON presenta dati null')

pd.options.display.max_rows = 100
pd.options.display.max_columns = 100

df[['Coderre', 'Bergeron', 'Joly']].sum().plot(kind='barh')
```

```python
# task 2:
'''• Con un grafico a barre confrontiamo il numero di votanti per ogni distretto'''

df[['district','total']].plot(kind='barh', x='district', y ='total', figsize=(12,12) )
```

```python
# task 3:
'''• Visualizzare un grafico a barre comparativo dove si confrontano
i voti presi nei primi 4 distretti per ogni candidato'''

df4 = df[['district','Coderre', 'Bergeron', 'Joly']].head(4)

df4.set_index('district').plot(kind='barh', figsize=(8, 6))
```
