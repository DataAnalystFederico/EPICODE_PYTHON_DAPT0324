```python
# task 1:
'''importa db stockdata e:
• Visualizziamo l'andamento delle azioni della Apple ("AAPL")'''

import pandas as pd

df = pd.read_csv('stockdata.csv')

# ricerca dati null e la loro percentuale
for column in df.columns:
    n_null = df[column].isna().sum()
    percentuale_null = round((n_null / df.shape[0]) * 100, 2)
    if percentuale_null > 0:
        print(column, 'contiene:', n_null,'dati null. Rappresentano il', percentuale_null,'% della colonna')

df.set_index('Date').AAPL.plot()
df.Date
```

```python
# task 2:
'''• Visualizziamo l'andamento di tutte le azioni del dataset'''

df.plot(x = 'Date', figsize = [12,5])
```

```python
# task 3:
'''• Visualizziamo l'andamento di Microsoft ("MSFT") e IBM ("IBM")
negli anni 2012 e 2013'''

df['Date'] = pd.to_datetime(df['Date'])
filtro_2012 = df.Date.dt.year == 2012
filtro_2013 = df.Date.dt.year == 2013

df.loc[filtro_2012 | filtro_2013, ['MSFT', 'IBM']].plot()
```

```python
# task 4:
'''Con un istogramma, troviamo il range di valori che 
le azioni di Microsoft hanno assunto più spesso'''

df.MSFT.plot(kind = 'hist')
```
