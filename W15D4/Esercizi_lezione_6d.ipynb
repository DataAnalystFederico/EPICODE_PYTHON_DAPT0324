```python
# task 1:
'''importa db automobile.csv:
• Ci sono valori nulli? Dove? Quanti?'''

import pandas as pd

df = pd.read_csv('automobile.csv')

pd.options.display.max_rows = 100
pd.options.display.max_columns = 1000

df.isna().sum()
print(df.isna().sum()) # si, in normalized-losses e num-of-doors, rispettivamente 37 e 2

# calcoliamo percentuale dati nulli su colonne in cui sono presenti
percentuale_null_normalized_losses = (df['normalized-losses'].isna().sum()) / (df.shape[0]) * 100
percentuale_null_normalized_losses

# sostituito temporaneamente il null con la media numerica della colonna
df['normalized-losses'].fillna(121)
```

```python
# task 2:
'''• Quali righe hanno un valore nullo nella colonna num-of-doors?'''

df[df['num-of-doors'].isnull()]
# oppure:
df[df['num-of-doors'].isna()].index

# isna = isnull
```

```python
# task 3:
'''• Esaminando i dati, cerchiamo dei valori da sostituire 
ai valori nulli nella colonna num-of-doors'''

df[df['num-of-doors'].isna()]

df.loc[26, 'num-of-doors'] = 'four'
df.loc[60, 'num-of-doors'] = 'four'

#df.loc[[26,60]]

df.head(29)
```

```python
# ricerca dati null e loro percentuale
for column in df.columns:
    n_null = df[column].isna().sum()
    percentuale_null = round((n_null / df.shape[0]) * 100, 2)
    if percentuale_null > 0:
        print(column, 'contiene:', n_null,'dati null. Rappresentano il', percentuale_null,'% della colonna')
```
