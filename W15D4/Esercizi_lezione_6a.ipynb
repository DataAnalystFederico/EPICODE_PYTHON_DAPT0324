```python
# task 3:
'''Creiamo un DataFrame sintetico, che contiene i guadagni mensili di diverse annate,
con il seguente codice:

years = 5 a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), 
"Anno": np.repeat(list(range(years)), 12), 
"Guadagni": np.random.randint(800, 5000, 12*years)})

• Calcola la somma cumulativa delle vendite per ogni mese utilizzando il metodo .cumsum()

• Come sopra, ma diviso per ogni anno'''

years = 5
a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), "Anno": np.repeat(list(range(years)), 12), 
"Guadagni": np.random.randint(800, 5000, 12*years)})

guadagni_cumulativi = a.Guadagni.cumsum()

guadagni_cumulativi_diviso_anni = guadagni_cumulativi / a['Anno']

print(guadagni_cumulativi, guadagni_cumulativi_diviso_anni)
```

```python
# task 4:
'''Dal database AdventureWorks estraiamo la tabella dimcustomer.

• Trasformiamo i nomi dei clienti in modo che abbiano solo lettere minuscole, 
e i cognomi in modo che abbiano solo lettere maiuscole'''

query1 = '''select * from dimcustomer'''
dimcustomer = pd.read_sql(query1, db_engine)

dimcustomer['nome_minuscolo'] = dimcustomer['FirstName'].str.lower() 
dimcustomer['cognome_maiuscolo'] = dimcustomer.LastName.str.upper()
dimcustomer.head()
```

```python
# task 5:
'''• Sulla colonna EmailAddress, utilizzando il metodo .str.split(), 
estraiamo nome utente e dominio'''

dimcustomer['nome_utente'] = dimcustomer.EmailAddress.str.split('@').str[0]
dimcustomer['dominio'] = '@' + (dimcustomer.EmailAddress.str.split('@').str[1])
dimcustomer
```

```python
# task 6:
'''• Sulla colonna Phone, estraiamo ogni parte del numero
(ad es. da "1 (11) 500 555-0162" a ["1", "(11)", "500", "555-0162"])'''

dimcustomer.Phone.str.split(' ')
```

```python
# task 7:
'''• Utilizzando il metodo .str.contains(), 
estraiamo tutti gli indirizzi e-mail che contengono il numero "21".'''

dimcustomer_email_21 = dimcustomer['EmailAddress'][dimcustomer.EmailAddress.str.contains('21')]
dimcustomer['email_21'] = dimcustomer['EmailAddress'][dimcustomer.EmailAddress.str.contains('21')]

print(dimcustomer_email_21)
dimcustomer
```

```python
# task 8:
'''• Estraiamo tutti gli indirizzi e-mail che contengono 
il numero "20" oppure il numero "10".'''

dimcustomer_email_20_10 = dimcustomer['EmailAddress'][dimcustomer.EmailAddress.str.contains('20') | dimcustomer.EmailAddress.str.contains('10')]
dimcustomer['email_20_10'] = dimcustomer_email_20_10

print(dimcustomer_email_20_10)
dimcustomer
```

```python
# task 9:
'''• Calcolare la lunghezza di ogni indirizzo e-mail ed estrarre
i dieci più lunghi e i dieci più corti'''

lunghezza_mail_ordinate = dimcustomer.EmailAddress.str.len().sort_values()
dieci_indirizzi_più_corti = dimcustomer.EmailAddress.iloc[lunghezza_mail_ordinate.index].head(10)
dieci_indirizzi_più_lunghi = dimcustomer.EmailAddress.iloc[lunghezza_mail_ordinate.index].tail(10)

print(dieci_indirizzi_più_corti, dieci_indirizzi_più_lunghi)
```

```python
# task 10:
'''• Modificare il dominio degli indirizzi e-mail da "adventure-works.com" 
a "aw-db.com" mediante il metodo .str.replace()'''

dimcustomer['indirizzo_mail_modificato'] = dimcustomer.EmailAddress.str.replace('@adventure-works.com', 'aw-db.com')
dimcustomer
```
