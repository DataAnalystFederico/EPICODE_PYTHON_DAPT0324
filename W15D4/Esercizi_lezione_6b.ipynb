```python
# task 1:
'''importa db facebook.csv:
• Con la funzione pd.to_datetime() convertiamo la colonna status_published in formato Timestamp'''

import pandas as pd

df = pd.read_csv('facebook.csv')

df['status_published'] = pd.to_datetime(df['status_published'])
df
```

```python
df.status_published.sample(5)
```

```python
# task 2:
'''• Utilizzando il metodo .dt.strftime(), formatta le date delle pubblicazioni 
in un formato specifico (ad esempio, "%Y-%m-%d %H:%M:%S").'''

df['status_published'] = df.status_published.dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df['status_published']
```

```python
# task 3:
'''• Utilizzando gli attributi .dt.year , .dt.month , .dt.day ,
.dt.dayofweek , .dt.dayofyear, ottieniamo informazioni specifiche sulle 
date delle transazioni, come l'anno, il mese, il giorno della settimana, il giorno dell'anno, ecc.'''

anno = df['status_published'].dt.year
print(anno)

mese = df['status_published'].dt.month
print(mese)

giorno = df['status_published'].dt.day
print(giorno)

giorno_settimana = df['status_published'].dt.dayofweek
print(giorno_settimana)

giorno_anno = df['status_published'].dt.dayofyear
print(giorno_anno)
```

```python
# task 4:
'''• Estraiamo solo i post relativi al 2012'''

status_2012 = df[df.status_published.dt.year == 2012]
status_2012[['status_id', 'status_published']]
```

```python
# task 5:
'''• Estraiamo solo i post relativi a maggio 2018'''

df_status_05_2018 = df[(df.status_published.dt.year == 2018) & (df.status_published.dt.month == 5)]
df_status_05_2018[['status_published', 'status_id']]
```

```python
# task 6:
'''Confrontiamo il numero di post pubblicati nei weekend 
rispetto al numero di post pubblicati nel resto della settimana'''

n_post_fine_settimana = df['status_id'][(df.status_published.dt.dayofweek == 5) | (df.status_published.dt.dayofweek == 6)].count()
n_posto_giorni_lavorativi = df['status_id'][(df.status_published.dt.dayofweek == 0) | (df.status_published.dt.dayofweek == 1)|(df.status_published.dt.dayofweek == 2)|(df.status_published.dt.dayofweek == 3)|(df.status_published.dt.dayofweek == 4)].count()
print(n_post_fine_settimana, n_posto_giorni_lavorativi)
```

```python
# task 7:
'''• Troviamo il primo e ultimo post pubblicati in ogni anno'''

df['anno'] = df['status_published'].dt.year

primo_post = df[['status_id','status_published', 'anno']].loc[df.groupby('anno')['status_published'].idxmin()]
ultimo_post = df[['status_id','status_published', 'anno']].loc[df.groupby('anno')['status_published'].idxmax()]

primo_post
```

```python
# task 8:
'''• Utilizzando il metodo .tz_localize(), aggiungiamo 
un fuso orario alle date delle transazioni'''

df['status_published'].dt.tz_localize('Europe/Rome')
```

```python
# task 9:
'''• Calcoliamo la differenza di tempo tra due transazioni'''
data_diff_primo_secono_indice = df.status_published.iloc[0] - df.status_published.iloc[1]
data_diff_primo_secono_indice
```

```python
# task 9:
'''• Quanti tipi di post ci sono? E quanti per ogni tipo?'''

n_status_type = df.status_type.value_counts().count()
n_post_per_tipo = df.status_type.value_counts()

print('ci sono:', n_status_type, 'tipi di stato')
n_post_per_tipo
```

```python
# ricerca dati null e loro percentuale
for column in df.columns:
    n_null = df[column].isna().sum()
    percentuale_null = (n_null / df.shape[0]) * 100
    if percentuale_null > 0:
        print(column, 'contiene:', n_null, ',', 'la percentuale di dati null in colonna è:', percentuale_null)
```
